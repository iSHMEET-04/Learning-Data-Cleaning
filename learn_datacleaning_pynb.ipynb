```python
for i in dataset.select_dtypes(include='object').columns:
  dataset[i].fillna(dataset[i].mode()[0], inplace=True)
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer(strategy='mean')
cols_to_impute=['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term',
       'Credit_History']
dataset[cols_to_impute]=si.fit_transform(dataset[cols_to_impute])
```

```python
dataset.isnull().sum()
```

```python
dataset.head(10)
```

```python
en_cols=dataset[['Gender','Married']]
en_cols.head(10)
```

```python
pd.get_dummies(en_cols)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one_hot=OneHotEncoder(drop='first')
arr=one_hot.fit_transform(en_cols).toarray()
```

```python
new=pd.DataFrame(arr,columns=[	'Gender_Male',		'Married_Yes'])
```

```python
dataset=pd.concat([dataset, new], axis=1)
```

```python
dataset.head(10)
```

```python
dataset.drop(columns=['Gender','Married'], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

```

```python
lab= LabelEncoder()
dataset['Property_Area']=lab.fit_transform(dataset['Property_Area'])
```

```python
dataset.head(10)
```

```python
dataset['Education'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
order= [['Not Graduate', 'Graduate']]
ord_en= OrdinalEncoder(categories=order)
```

```python
dataset['Education']=ord_en.fit_transform(dataset[['Education']])
```

```python
dataset.head(10)
```

```python
dataset.describe()
```

```python
sns.boxplot(x='CoapplicantIncome',data=dataset)
```

```python
q1= dataset['CoapplicantIncome'].quantile(0.25)
q3= dataset['CoapplicantIncome'].quantile(0.75)
iqr=q3-q1

min_range= q1 - (1.5*iqr)
max_range= q3 + (1.5*iqr)

```

```python
dataset= dataset[dataset['CoapplicantIncome']<= max_range]
```

```python
dataset.shape
```

```python
dataset['LoanAmount'].describe()
```

```python
min_rangee = dataset['LoanAmount'].mean() - (3* dataset['LoanAmount'].std())
max_rangee = dataset['LoanAmount'].mean() + (3* dataset['LoanAmount'].std())

dataset= dataset[dataset['LoanAmount']<= max_rangee]
dataset= dataset[dataset['LoanAmount']>= min_rangee]
```

```python
sns.boxplot(x='LoanAmount', data=dataset)
```

```python
z_score= (dataset['LoanAmount']-dataset['LoanAmount'].mean())/dataset['LoanAmount'].std()
print(z_score)
```

```python
dataset['z_score']=z_score
(dataset['z_score'])<3
```

```python
sns.boxplot(x='LoanAmount', data=dataset)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss= StandardScaler()
ss.fit(dataset[['ApplicantIncome']])
```

```python
dataset['ApplicantIncome_ss']=pd.DataFrame( ss.transform(dataset[['ApplicantIncome']]), columns=['x'])
```

```python
dataset.head()
```

```python
plt.figure(figsize=(4,4))
plt.subplot(1,2,1)
plt.title('Before')
sns.distplot(dataset['ApplicantIncome'])

plt.subplot(1,2,2)
plt.title('After')
sns.distplot(dataset['ApplicantIncome_ss'])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms= MinMaxScaler()
mms.fit(dataset[['CoapplicantIncome']])
```

```python
dataset['CoapplicantIncome_mms']= pd.DataFrame(mms.transform(dataset[['CoapplicantIncome']]), columns=['x'])
```

```python
plt.figure(figsize=(4,4))
plt.subplot(1,2,1)
plt.title('Before')
sns.distplot(dataset['CoapplicantIncome'])

plt.subplot(1,2,2)
plt.title('After')
sns.distplot(dataset['CoapplicantIncome_mms'])
```

```python
dataset.shape
```

```python
dataset.drop_duplicates(inplace=True)
dataset.shape
```

```python
dataset['Dependents'].replace('3+',3, inplace=True)
dataset['Dependents']=dataset['Dependents'].astype('int64')
```

```python
dataset.info()
```

```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np
```

```python
ft= FunctionTransformer(func=np.log1p)
```

```python
dataset['CoapplicantIncome_ft']=ft.fit(dataset[['CoapplicantIncome']])
dataset['CoapplicantIncome_ft']=ft.transform(dataset[['CoapplicantIncome']])
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(dataset['CoapplicantIncome'])
plt.title('Before Function Transform')

plt.subplot(1,2,2)
sns.distplot(dataset['CoapplicantIncome_ft'])
plt.title('After Function Transform')
```
